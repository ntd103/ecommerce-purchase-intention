```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('../data/online_shoppers_intention.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(f"Số lượng hàng bị trùng lặp: {df.duplicated().sum()}")
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='Revenue', data=df)
plt.title('Phân bổ số phiên có và không có doanh thu')
plt.show()
```

```python
print("Tỷ lệ các lớp trong cột 'Revenue':")
print(df['Revenue'].value_counts(normalize=True))
```

```python
# Phân tích các biến phân loại khác
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
```

```python
# Biểu đồ cho VisitorType
sns.countplot(ax=axes[0], x='VisitorType', data=df, order=df['VisitorType'].value_counts().index)
axes[0].set_title('Phân bổ loại khách truy cập')
```

```python
sns.countplot(ax=axes[1], x='Month', data=df, order=df['Month'].value_counts().index)
axes[1].set_title('Lượng truy cập theo tháng')
plt.sca(axes[1])
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
